## Exploring Categorical Variables

<b>NOTE:</b> To get the exact value counts for each category of a variable, uncomment the first line of each cell.

```python
#df_raw["food_category"].value_counts()

(p9.ggplot(data=df_raw,
           mapping=p9.aes(x='food_category'))
    + p9.geom_bar(fill = "#FF6660")
    + p9.theme_bw()
    + p9.theme(axis_text_x = p9.element_text(angle=90))
)
```
